```python
# Loading libraries
import pandas as pd
import numpy as np
```

```python
# Global variables and constants
DATA_FILENAME = 'OnlineNewsPopularity.csv'
```

```python
data = pd.read_csv(DATA_FILENAME, index_col=False)
```

```python
data.columns = data.columns.str.replace(" ", "")
```

```python
data['log_shares'] = np.log1p(data['shares'])
```

```python
x_cols = data.columns[1:-2]
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data[x_cols], data['log_shares'], test_size=0.33)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
import xgboost as xgb
```

```python
train_xgb = xgb.DMatrix(X_train, y_train)
test_xgb = xgb.DMatrix(X_test, y_test)
```

```python
params = {'objective': 'reg:linear'}
num_rounds = 200
```

```python
model = xgb.train(params, train_xgb, num_rounds)
```

```python
predictions = model.predict(test_xgb)
```

```python
predictions = np.expm1(predictions)
```

```python
plt.rcParams['figure.figsize'] = (10, 10)
xgb.plot_importance(model)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(np.expm1(y_test), predictions)
```

```python
def root_mean_squared_error(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))
```

```python
root_mean_squared_error(np.expm1(y_test), predictions)
```
